```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#loading the dataset and checking the first few rows
df = pd.read_csv(r"C:\Users\sanke\Downloads\co2 Emission Africa.csv")
df.head(5)
```

```python
#basic information of the dataset
df.info()

print("\nsummary statisics :")
df.describe()
```

```python
#data cleaning - handling missing values and duplicates and dropping null values

df.isnull().sum()

df.duplicated().sum()

df.dropna()
```

```python
#top 10 emiitter countries across all years 
top_emitters = df.groupby('Country')['Total CO2 Emission including LUCF (Mt)'].sum().sort_values(ascending=False).head(10)
top_emitters_df = top_emitters.reset_index().rename(columns={'Total CO2 Emission including LUCF (Mt)': 'Total CO2 Emission excluding LUCF(Mt)_sum'})
print("TOP 10 EMITTERS COUNTRIES ACROSS ALL YEARS")
display(top_emitters_df)

#bar plot 
plt.figure(figsize = (10,5))
sns.barplot(data=top_emitters_df, x='Country', y='Total CO2 Emission excluding LUCF(Mt)_sum')
plt.xticks(rotation=45)
plt.title('Top 10 CO2 Emitting Countries in Africa (All Years)')
plt.tight_layout()
```

```python
#checking the year with highest emissions on average
# Group data by year and compute average total emissions
yearly_avg = (
    df.groupby('Year')['Total CO2 Emission excluding LUCF (Mt)']
    .mean()
    .reset_index()
    .rename(columns={'Total CO2 Emission excluding LUCF (Mt)': 'Average_Emission'})
)

# Find the year with the maximum average emission
max_row = yearly_avg.loc[yearly_avg['Average_Emission'].idxmax()]
print(f"Year with the highest average CO emission: {int(max_row['Year'])}")
print(f"Average Emission: {max_row['Average_Emission']:.2f} Mt")

# Plot the trend over years
plt.figure(figsize=(10,5))
sns.lineplot(data=yearly_avg, x='Year', y='Average_Emission', marker='o')
plt.title('Average CO Emissions Across Years in Africa')
plt.xlabel('Year')
plt.ylabel('Average CO Emission (Mt)')
plt.grid(True)
plt.show()
```

```python
#trends over time  for a single country chosen by the user
country = input("Enter a country name: ")
if country not in df['Country'].unique():
    print("country not found in dataset")
else:
    country_df = df[df['Country'] == country].sort_values('Year')
    plt.figure(figsize=(10,5))
    sns.lineplot(data=country_df, x='Year', y='Total CO2 Emission excluding LUCF (Mt)', marker='o')
    plt.title(f'CO Emissions Over Time in {country}')
    plt.ylabel('Total CO Emission (Mt)')
    plt.xlabel('Year')
```

```python
#Detect sector-like columns
sector_keywords = ['energy', 'transport', 'industrial', 'industry', 'agri', 'agriculture', 'waste', 'electricity', 'building', 'bunker', 'fugitive']
cols = df.columns.tolist()
sector_cols = [c for c in cols if any(k in c.lower() for k in sector_keywords)]
print('Detected sector columns:', sector_cols)

# Detect total emission column
total_candidates = [c for c in cols if 'total' in c.lower() and 'exclud' in c.lower()]
if not total_candidates:
    total_candidates = [c for c in cols if 'total' in c.lower()]
if not total_candidates:
    print('No total emission column detected automatically. Please inspect df.columns.')
else:
    total_col = total_candidates[0]
    print('Using total column:', total_col)
    df['Sector_Sum'] = df[sector_cols].sum(axis=1, skipna=True)
    df['Sector_vs_Total_diff'] = df[total_col] - df['Sector_Sum']
    display(df['Sector_vs_Total_diff'].describe())

    plt.figure(figsize=(7,6))
    sns.scatterplot(x='Sector_Sum', y=total_col, data=df)
    maxv = max(df['Sector_Sum'].max(), df[total_col].max())
    plt.plot([0, maxv], [0, maxv], linestyle='--', color='red')
    plt.xlabel('Sum of sector emissions (Mt)')
    plt.ylabel(f'Reported {total_col}')
    plt.title('Sector Sum vs Reported Total Emissions (excl. LUCF if present)')
    plt.tight_layout()
    plt.show()
```

```python
# --- Total emissions by Sub-Region ---
subregion_candidates = [c for c in df.columns if 'sub' in c.lower()]
if not subregion_candidates:
    print('No Sub-Region column detected automatically. Inspect df.columns to find it (e.g., "Sub-Region").')
else:
    sub_col = subregion_candidates[0]
    total_candidates = [c for c in df.columns if 'total' in c.lower() and 'exclud' in c.lower()]
    if not total_candidates:
        total_candidates = [c for c in df.columns if 'total' in c.lower()]
    total_col = total_candidates[0]
    region_emissions = df.groupby(sub_col)[total_col].sum().sort_values(ascending=False).reset_index()
    plt.figure(figsize=(10,6))
    sns.barplot(data=region_emissions, x=sub_col, y=total_col)
    plt.xticks(rotation=30)
    plt.title('Total CO2 Emissions by Sub-Region (sum across years and countries)')
    plt.ylabel('Total CO2 Emission (Mt)')
    plt.xlabel('Sub-Region')
    plt.tight_layout()
    plt.show()
    print('\nTop emitting region:', region_emissions.iloc[0,0])
    display(region_emissions)
```

```python
# --- Sector breakdown for a user-selected country and year ---
country = input('Enter a country name (case-insensitive, e.g., Kenya): ').strip()
year_input = input('Enter a year (e.g., 2015): ').strip()

try:
    year_sel = int(year_input)
except:
    print('Invalid year. Please enter an integer like 2015.')
    year_sel = None

if country=='' or year_sel is None:
    print('Please provide both a country and a valid year.')
else:
    matches = df['Country'].unique()
    match = None
    for m in matches:
        if m.lower() == country.lower():
            match = m
            break
    if match is None:
        print(f'Country "{country}" not found. Sample of countries:', list(matches)[:10])
    else:
        sector_keywords = ['energy', 'transport', 'industrial', 'industry', 'agri', 'agriculture', 'waste', 'electricity', 'building', 'bunker', 'fugitive']
        cols = df.columns.tolist()
        sector_cols_local = [c for c in cols if any(k in c.lower() for k in sector_keywords)]
        total_candidates = [c for c in cols if 'total' in c.lower() and 'exclud' in c.lower()]
        if not total_candidates:
            total_candidates = [c for c in cols if 'total' in c.lower()]
        if not total_candidates:
            print('No total emission column found. Cannot proceed.')
        else:
            total_col = total_candidates[0]
            subset = df[(df['Country']==match) & (df['Year']==year_sel)]
            if subset.empty:
                print('No data for', match, 'in year', year_sel)
            else:
                row = subset.iloc[0]
                sector_values = {s: row[s] for s in sector_cols_local if s in subset.columns}
                if not sector_values:
                    print('No sector columns available for this dataset.')
                else:
                    sec_df = pd.DataFrame.from_dict(sector_values, orient='index', columns=['CO2']).reset_index().rename(columns={'index':'Sector'})
                    sec_df = sec_df.sort_values('CO2', ascending=False)
                    print(f"\nSector contributions for {match} in {year_sel} (showing top sectors):\n")
                    display(sec_df)
                    plt.figure(figsize=(8,5))
                    sns.barplot(data=sec_df, x='CO2', y='Sector')
                    plt.title(f'Sector CO2 contributions for {match} in {year_sel}')
                    plt.xlabel('CO2 Emissions (Mt)')
                    plt.tight_layout()
                    plt.show()
                    top_sector = sec_df.iloc[0]
                    print(f"Top sector: {top_sector['Sector']} with {top_sector['CO2']:.2f} Mt")
```
